```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Data.xlsx")

df.head()
df.info()
df.describe()
```

```python
# انتخاب ستون‌های عددی برای تحلیل
numeric_columns = ["فی", "مبلغ ناخالص فروش", "مبلغ خالص فروش", "مبلغ کل", "مقدار فروش", "مقدار فروش خالص"]

# رسم pairplot برای نمایش روابط بین متغیرهای عددی
sns.pairplot(df[numeric_columns], diag_kind='kde')
plt.show()
```
